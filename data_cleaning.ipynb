```python
import urllib
from urllib.request import urlopen
import pandas as pd
import numpy as np
```

```python
def make_soup(url):
    req = urllib.request.Request(url, 
    headers={'User-Agent': 'Mozilla/5.0'})
    webpage = urlopen(req)
    main_doc = webpage.read()
    return main_doc
```

```python
#prospect names 
soup= make_soup("https://www.rookiescale.com/2022-consensus-board/")
namesDF=pd.read_html(soup)[0]
```

```python
prospects = namesDF['Prospect'].tolist()
#add hypen to each player (names include hypen in URL)
prospects_list = list(map(lambda x: x.replace(' ', '-'), prospects))
#convert names to lowercase
prospects_list = list(map(lambda x: x.lower(), prospects_list))
```

```python
df_list2=[]
#get college stats for players in NBA draft, each have own URL
for prospect in prospects_list:
    #print(player)
    if prospect.endswith('-jr.') == True:
        prospect=prospect.replace('-jr.', 'jr')
    for i in range(8):
        try:
            soup= make_soup("https://www.sports-reference.com/cbb/players/"+prospect+"-"+str(i)+".html")
            temp_df=pd.read_html(soup)[0]
            temp_df["name"] = prospect #add player name to dataframe
            print(prospect+" MADE IT")
            df_list2.append(temp_df)
        except Exception as e:
                print(prospect+" MISSING")
                pass
```

```python
years = [2010, 2011, 2012, 2013 ,2014 ,2015 ,2016 ,2017, 2018, 2019, 2020, 2021]
df_list=[]

#get draft data from 2010-2021
for year in years:
    #print(team)
    try:
        soup= make_soup("https://www.basketball-reference.com/draft/NBA_"+str(year)+".html/")
        df_list.append(pd.read_html(soup)[0])
    except Exception as e:
            pass
```

```python
#store data in df
df = pd.concat(df_list)
```

```python
df.head()
```

```python
colnames=["Rk",	"Pk",	"Tm",	"Player",	"College"	,"Yrs",	"G"	,"MP",	"PTS",	"TRB",	"AST",	"FG%",
	"3P%",	"FT%"	,"MP"	,"PTS",	"TRB",	"AST",	"WS",	"WS_48",	"BPM",	"VORP"]
```

```python
#change column names
df.columns = colnames
```

```python
df.head()
```

```python
#create a list from Player column
players = df['Player'].tolist()
```

```python
#add hypen to each player (names include hypen in URL)
player_list = list(map(lambda x: x.replace(' ', '-'), players))
```

```python
#convert names to lowercase
player_list = list(map(lambda x: x.lower(), player_list))
```

```python
df_list2=[]
#get college stats for players in NBA draft, each have own URL
for player in player_list:
    #print(player)
    try:
        soup= make_soup("https://www.sports-reference.com/cbb/players/"+player+"-1.html")
        temp_df=pd.read_html(soup)[0]
        temp_df["name"] = player #add player name to dataframe
        df_list2.append(temp_df)
    except Exception as e:
            pass
```

```python
college = pd.concat(df_list2)
```

```python
#remove hypen from name
college['name'] = college['name'].str.replace('-', ' ')
```

```python
#convert names in draft df to lower case so we can join it with college dataframe
df['Player'] = df['Player'].str.lower()
```

```python
#inner join draft dataframe with college dataframe on the college and player name 
#add _college suffix to college variables to avoid duplicate variable names
df =  df.merge(college, left_on=["Player","College"], right_on=["name","School"], how='inner',suffixes=('','_college'))
```

```python
#get rid of NBA stats 
df.drop(['Tm',
 'Player',
 'College',
 'Yrs',
 'G',
 'MP',
 'PTS',
 'TRB',
 'AST',
 'FG%',
 '3P%',
 'FT%',
 'MP',
 'PTS',
 'TRB',
 'AST',
 'WS',
 'WS_48',
 'BPM',
 'VORP'],inplace=True,axis=1)
```

```python
#keep only career college stats
#remove individual seasons  
df_clean = df[df['Season'] == "Career"]
```

```python
df_clean.columns
college.columns
```

```python
#drop unwanted cols
df_clean.drop(["Conf",'Unnamed: 27','Unnamed: 25'],inplace=True,axis=1)
```

```python
df_clean.drop("Season",inplace=True,axis=1)
df_clean.isna().sum()
```

```python
#convert cols to numeric 
df_clean[['Rk', 'Pk', 'G_college', 'GS', 'MP_college', 'FG', 'FGA',
       'FG%_college', '2P', '2PA', '2P%', '3P', '3PA', '3P%_college', 'FT',
       'FTA', 'FT%_college', 'ORB', 'DRB', 'TRB_college', 'AST_college', 'STL',
       'BLK', 'TOV', 'PF', 'PTS_college', 'SOS']] = df_clean[['Rk', 'Pk', 'G_college', 'GS', 'MP_college', 'FG', 'FGA',
       'FG%_college', '2P', '2PA', '2P%', '3P', '3PA', '3P%_college', 'FT',
       'FTA', 'FT%_college', 'ORB', 'DRB', 'TRB_college', 'AST_college', 'STL',
       'BLK', 'TOV', 'PF', 'PTS_college', 'SOS']].apply(pd.to_numeric)
```

```python
#create lottery varible 
#if player is a top 14 pick then they are in the lottery 
df_clean['lottery'] = np.where(df_clean['Pk']<=14, True, False)
```
